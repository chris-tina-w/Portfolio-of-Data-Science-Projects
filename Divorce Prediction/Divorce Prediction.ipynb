# Divorce Prediction 

_This project was conducted during my data mining university course using the Data Mining software Weka._

Data mining methods were employed to identify whether there are certain divorce predictors which could be identified early in married couples. The data used is from 170 individuals (86 married and 84 divorced) who have filled out surveys with the Divorce Predictors Scale questions which predicts the divorce rates of couples by 98.23% accuracy.

Within the survey, features which could be identified as divorce preditors include perceptions of value systems, attitudes about trust, satisfaction with sexuality, etc. In our data, the features have been de-identified as nominal attributes.

In this study, the success of the Divorce Predictors Scale was investigated using Multilayer Perceptron Neural Network, C4.5 Decision tree algorithm and logistic regression. I also compare the findings from my model to the ones from the study _Divorce Prediction Using Correlation Based Feature Selection and Artificial Neural Networks_ by Yontem et al. (2019).

The accuracy for is higher for 10-fold cross validation than one 70:30 train:test split for all the tests across the four datasets, with the exception of logistic regression results in two of the datasets (all 54 attributes and 6 most correlated attributes). This is because the 10-fold cross validation method is a more robust way of training the data as it performs 10 iterations of the training and testing data subsets with different splits of the subsets each time. Whereas the 70:30 train:test split is a single round of testing the training and testing sets in a 70% to 30% split. Therefore, in most cases the 10-fold method would produce higher accuracy and is a more suitable method. 

As seen in the results table, the attributes proposed by Yontem et al.(2019) produced slightly higher accuracy (10-fold cross valid.) and kappa (10-fold cross valid.) across all three classification techniques and also for 70:30 train:test split results for Neural Network (MLP). The 6 most correlated attributes produced slightly higher accuracy and kappa for logistic regression test results in 70:30 train:test split. Lastly, the accuracy and kappa results were the same for Random Forest in the 70:30 train:test split. The slight difference could suggest that the authors choose a more significant set of 6 features which affect divorce based on potentially different parameters used in their correlation-based feature selection method. 

Multilayer perceptron neural network and random forest produced the same values for accuracy and kappa for both datasets with the 10-fold cross validation. This suggests that both methods are robust and produces stable results without being distracted by potential noise. 

Considering the different datasets:
- All 54 attributes dataset: RF model performed best in accuracy and kappa and had fastest build time
- 6 most correlated attributes: Logistic Regression model performed best in accuracy and kappa and had fastest build time
- 6 least correlated attributes: Multilayer Perceptron Neural Network had the the highest accuracy and kappa for 10-fold cross validation but longest build time. This model would only be chosen if accuracy was prioritised.
- Yontem et al. dataset: Logistic Regression performed best in accuracy and had the fastest build time
