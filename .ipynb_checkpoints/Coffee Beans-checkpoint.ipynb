# Regression Model for Coffee Bean Prices

With the recent wholesale price of coffee beans reaching a record high in December 2024, consumers are uneasy that the price of a coffee will skyrocket to $10-12 a cup. The increasing cost of this commodity is associated with recent unfavourable growing conditions as well as additional geopolitical factors impacting supply chains. Most of the coffee we drink in Australia is imported, with locally grown beans making up less than 1% of the market. 
 
I wanted to look into how well closing coffee bean prices could be predicted with regression models using the open, high and low prices as well as the volume. The data that I use for this project is compiled by Tesyon Hwase (Department of Software Engineering, Nankai University, Ethiopia) from data collected from the Ethiopian Commodity Market looking at prices between 2012 and 2018. 

The dataset contains 1540 instances and 7 attributes.

References: https://data.mendeley.com/datasets/c55sbp7dgv/1 (dataset)

```python
# split into feature and target variables
X = df.drop(columns=['Close'])
y = df['Close']
```

```python
# split into training, validation and test sets
val_mask = (df.index.year == 2017)
test_mask = (df.index.year == 2018)

X_train = X[~(val_mask | test_mask)]  
y_train = y[~(val_mask | test_mask)] 

X_val = X[val_mask]               
y_val = y[val_mask]  

X_test = X[test_mask]  
y_test = y[test_mask] 

# Print information about the splits
print(f"# of Training samples: {len(X_train)}")
print(f"# of Validation samples: {len(X_val)}")
print(f"# of Test samples: {len(X_test)}")

print(f"Min Date in Train: {X_train.index.min()} | Max Date in Train: {X_train.index.max()}")
print(f"Min Date in Validation: {X_val.index.min()} | Max Date in Validation: {X_val.index.max()}")
print(f"Min Date in Test: {X_test.index.min()} | Max Date in Test: {X_test.index.max()}")
```

```python
# function to train and evaluate the models
def train_and_evaluate_model(model, X_train, y_train, X_val, y_val, X_test, y_test, model_metrics, scale_features=True):

    # for scaling the features
    if scale_features:
        scaler = StandardScaler()
        X_train = scaler.fit_transform(X_train)
        X_val = scaler.transform(X_val)
        X_test = scaler.transform(X_test)

    # training the model
    model.fit(X_train, y_train)

    # predict on validation and test sets
    y_val_pred = model.predict(X_val)
    y_test_pred = model.predict(X_test)

    # compute metrics for validation and test set
    val_mae = mean_absolute_error(y_val, y_val_pred)
    val_mse = mean_squared_error(y_val, y_val_pred)
    val_rmse = np.sqrt(val_mse)
    test_mae = mean_absolute_error(y_test, y_test_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)
    test_rmse = np.sqrt(test_mse)

    # compute bias
    val_bias = np.mean(y_val_pred - y_val)
    test_bias = np.mean(y_test_pred - y_test)

    # print results
    print(f"Model: {model.__class__.__name__}")
    print(f"Validation MAE: {val_mae}, RMSE: {val_rmse}, MSE: {val_mse}, Bias: {val_bias}")
    print(f"Test MAE: {test_mae}, RMSE: {test_rmse}, MSE: {test_mse}, Bias: {test_bias}")

    # return metrics and model name as a dictionary
    model_metrics[model.__class__.__name__] = {
        "validation": {"MAE": val_mae, "RMSE": val_rmse, "MSE": val_mse, "Bias": val_bias},
        "test": {"MAE": test_mae, "RMSE": test_rmse, "MSE": test_mse, "Bias": test_bias},
    }

    return y_val_pred, y_test_pred, model_metrics

model_metrics = {}
```

```python
def plot_actual_vs_predicted(model, y_test, y_test_pred, xlabel="Date", ylabel="Coffee Prices"):
    plt.figure(figsize=(12,6))
    plt.plot(y_test.index, y_test.values, label="Actual", color="blue")
    plt.plot(y_test.index, y_test_pred, label="Predicted", color="red", linestyle="--")
    plt.title(f"Actual vs Predicted for {model}")
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.show()
```

### Linear Regression

```python
linear_model = LinearRegression()

# train and evaluate
y_val_pred_lr, y_test_pred_lr, model_metrics = train_and_evaluate_model(
    model=linear_model,
    X_train=X_train,
    y_train=y_train,
    X_val=X_val,
    y_val=y_val,
    X_test=X_test,
    y_test=y_test,
    model_metrics=model_metrics
)
```

```python
# visualise actual vs predicted prices
plot_actual_vs_predicted(
    model=linear_model,
    y_test=y_test,
    y_test_pred=y_test_pred_lr
)
```

### Ridge Regression

```python
ridge_model = RidgeCV()

# train and evaluate
y_val_pred_ridge, y_test_pred_ridge, model_metrics = train_and_evaluate_model(
    model=ridge_model,
    X_train=X_train,
    y_train=y_train,
    X_val=X_val,
    y_val=y_val,
    X_test=X_test,
    y_test=y_test,
    model_metrics=model_metrics
)
```

```python
# visualise actual vs predicted prices
plot_actual_vs_predicted(
    model=ridge_model,
    y_test=y_test,
    y_test_pred=y_test_pred_ridge
)
```

### Decision Tree

```python
tree_model = DecisionTreeClassifier(max_depth=50, random_state=42)

# train and evaluate
y_val_pred_tree, y_test_pred_tree, model_metrics = train_and_evaluate_model(
    model=tree_model,
    X_train=X_train,
    y_train=y_train,
    X_val=X_val,
    y_val=y_val,
    X_test=X_test,
    y_test=y_test,
    model_metrics=model_metrics
)
```

```python
# visualise actual vs predicted prices
plot_actual_vs_predicted(
    model=tree_model,
    y_test=y_test,
    y_test_pred=y_test_pred_tree
)
```

### Random Forest

```python
forest_model = RandomForestRegressor(n_estimators=100, random_state=42)

# train and evaluate
y_val_pred_forest, y_test_pred_forest, model_metrics = train_and_evaluate_model(
    model=forest_model,
    X_train=X_train,
    y_train=y_train,
    X_val=X_val,
    y_val=y_val,
    X_test=X_test,
    y_test=y_test,
    model_metrics=model_metrics
)
```

```python
# visualise actual vs predicted prices
plot_actual_vs_predicted(
    model=forest_model,
    y_test=y_test,
    y_test_pred=y_test_pred_forest
)
```

```python
# convert dictionary to DataFrame
formatted_table = []

for model, metrics in model_metrics.items():
    formatted_table.append({
        "Algorithm": model,
        "MAE": metrics["test"]["MAE"],
        "MSE": metrics["test"]["MSE"],
        "Bias": metrics["test"]["Bias"],  # Fix the key to lowercase 'bias'
    })

# create a DataFrame
df_metrics_table = pd.DataFrame(formatted_table)
```

### Conclusion

The Linear Regression model demonstrated the best performance on the test dataset among all evaluated algorithms. It achieved the lowest Mean Absolute Error (MAE) of approximately 1.52, indicating that, on average, its predictions deviated from the actual coffee bean prices by around 2 units. This model also had the lowest Mean Squared Error (MSE) and relatively low bias compared to other models, suggesting both higher accuracy and consistency in its predictions. 

Ridge regression with cross-validation has a slightly higher error than LinearRegression, but the bias is negative, indicating it underpredicts prices slightly on average. The regularisation may have prevented overfitting. The cross-validation in RidgeCV selects the best regularization parameter (alpha), but this added complexity might slightly increase the error compared to simple Linear Regression. 

Decision trees are better suited for classification tasks, not continuous regression. Random Forest combines multiple decision trees through bootstrap aggregation (bagging), which reduces the variance of individual trees. Random Forest is better when the relationships are highly non-linear or there are strong interactions between features, which might not be the case here.
